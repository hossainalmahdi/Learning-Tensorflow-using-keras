```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
#Initializing List to store data's
train_labels = [] 
train_samples = []
```

```python
for i in range(50):
    #creating data for younger individual who are in age (13-64)
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1) #append 1 is for having side effects
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0) #append 0 is for having NO side effects

for i in range(1000):
    #creating data for younger individual who are in age (13-64)
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0) #append 0 is for having no side effects
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1) #append 1 is for having side effects
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
#making numpy array for the fit function
train_labeles = np.array(train_labels)
train_samples = np.array(train_samples)
train_labeles,train_samples = shuffle(train_labels,train_samples)
```

```python
scaler = MinMaxScaler(feature_range =(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# NOW WE WILL TRY TO USE SEQUENTIAL MODEL & BUILD ARTIFICIAL NURAL NETWORK

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print("GPU AVAILABLE: ", len(physical_devices))
tf.config.experimental.set_memory_growth(physical_devices[0],True)
```

```python
model = Sequential([
    Dense(units=16, input_shape =(1,), activation= 'relu'),
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')
]
)
```

```python
model.summary()
```

# train network

```python
model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
)
```

# now its ready for training

```python
x=np.array(scaled_train_samples)
y=np.array(train_labeles)
```

```python
model.fit(x,y,batch_size=10,epochs=30,verbose=2,validation_split=0.1 )
#batch size= how many samples are included in one batch
#shuffle set to deafult always//verbose is to show the datas,parameter(0,1,2)
#validation split is basically keeps 10% data separate from the actual training data to check the validation of that data. if validation doesnt match with the accuracy or loss it becomes goes to underfit or overfit
```

# Confusion Matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels,y_pred=rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects','had_side_effects']
plot_confusion_matrix(cm=cm,classes=cm_plot_labels,title='VISUAL REPRESENTATION')
```
